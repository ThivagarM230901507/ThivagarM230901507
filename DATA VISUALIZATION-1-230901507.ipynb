```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Xpoints=np.array([0,10])
Ypoints=np.array([20,100])
print(plt.plot(Xpoints,Ypoints))
```

```python
x = np.array([0,5]) 
y = np.array([0,25]) 
print(plt.plot(x,y,marker='d')) 
```

```python
xpoints = np.array([1,2,6,8]) 
ypoints = np.array([3,8,1,10]) 
print(plt.plot(xpoints,ypoints)) 
```

```python
xpoints = np.array([0,1,2,3,4,5,6]) 
ypoints = np.array([2,-2,2,-2,2,-2,2]) 
print(plt.plot(xpoints,ypoints)) 
```

```python
x=np.arange(0,2*(np.pi),0.1)
y=np.sin(x)
print(plt.plot(x,y))
```
